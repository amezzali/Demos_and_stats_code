```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io.wavfile import write

%matplotlib inline

f = 440
sampleRate = 44100
duration = 10
sample = sampleRate*duration

x = np.arange(sample)

y = np.sin(2*np.pi*f*x/sampleRate)*10*x

scaled = np.int16(y/np.max(np.abs(y)) * 32767)
write('test.wav', 44100, scaled)


xPlot = np.arange(0,duration/f,0.0001)
yPlot = np.sin(2*np.pi*f*xPlot)*10*xPlot

plt.plot(xPlot,yPlot)
```
